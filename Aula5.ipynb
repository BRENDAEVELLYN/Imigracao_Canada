```python
# prompt: pesquisar os top 10 países com imigrantes para o Canadá

top_10 = df.sort_values(by='Total', ascending=False).head(10)
```

```python
# prompt: pesquisar qual pais no ano de 2011 imigrou mais para o Canada

df['2011'].idxmax()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
ax = sns.barplot(data=top_10, y=top_10.index, x='Total',orient='h')
ax.set(
    title="Top 10 países com mais imigrantes para o Canadá",
    xlabel="Total de Imigrantes",
    ylabel=''
)




plt.show()
```

```python
def gerar_grafico_palleta(pallete):
  fig,ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total',orient='h', palette=pallete)
  ax.set_title("Top 10 países com mais imigrantes para o Canadá",loc='left',fontsize=16)
  ax.set_xlabel("Total de Imigrantes")
  ax.set_ylabel('')

  plt.show()
```

```python
gerar_grafico_palleta('rocket')
```
